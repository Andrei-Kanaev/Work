1.3 Напишите функцию, которая по `username` ищет человека и выводит информацию о нем. Если пользователь с заданным `username` отсутствует, возбудите исключение `ValueError`

```python
import json


def find_user_by_username(username):
    with open(r"C:\юпитер ноутбук\WPy64-31050\notebooks\ТОД лабораторные\лаб 4\contributors_sample.json", "r") as f:
        data = json.load(f)

    for user in data:
        if user.get("username") == username:
            return user

    raise ValueError("User with username {} not found".format(username))


try:
    user_info = find_user_by_username("uhebert")
    print(json.dumps(user_info, indent=4))
except ValueError as e:
    print(e)
```

### pickle

```python
import json

with open(r"C:\юпитер ноутбук\WPy64-31050\notebooks\ТОД лабораторные\лаб 4\contributors_sample.json", "r") as f:
    data = json.load(f)

positions = {}

for user in data:
    position = user.get("position")
    username = user.get("username")
    if position in positions:
        positions[position].append(username)
    else:
        positions[position] = [username]

print(positions)
```

2.2 Сохраните результаты в файл `job_people.pickle` и в файл `job_people.json` с использованием форматов pickle и JSON соответственно. Сравните объемы получившихся файлов. При сохранении в JSON укажите аргумент `indent`.

```python
import pickle
import json
import os

with open("job_people.pickle", "wb") as f:
    pickle.dump(positions, f)
with open("job_people.json", "w") as f:
    json.dump(positions, f, indent=4)

print(f"job_people.pickle: {os.path.getsize('job_people.pickle')} bytes")
print(f"job_people.json: {os.path.getsize('job_people.json')} bytes")
    
```
